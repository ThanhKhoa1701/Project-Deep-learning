```python
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
```

# Load Data

```python
(X_train, y_train), (X_test,y_test)=keras.datasets.fashion_mnist.load_data()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
class_names=['T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle boot']
class_names[y_train[0]]

class_nums = len(class_names)
```

```python
def plot_data(x_data: np.ndarray, y_data: np.ndarray):
  nrows,ncols = 2,5
  fig,axes=plt.subplots(nrows,ncols,figsize=(8,4))

  len_x = x_data.shape[0]
  for idx in range(nrows*ncols):
    ax = axes[idx // ncols, idx % ncols]

    img_idx = random.randint(0, len_x)

    ax.imshow(x_data[img_idx], cmap = 'gray')
    ax.set(xticks = [] , yticks = [])

    true_label_str = f"True: {class_names[y_data[img_idx]]}"
    ax.set_title(true_label_str, color = 'green')
  plt.tight_layout()
  plt.show()

plot_data(X_train, y_train)
```

# Image Pre-Processing

```python
# chuyển sang kiểu dữ liệu float32 và scale ảnh vào vùng từ [0,1]
X_train = X_train.astype(np.float32) / 255
X_test = X_test.astype(np.float32) / 255
```

```python
X_test.shape, X_test.dtype
```

```python
# expand  channel shape
X_train = np.expand_dims(X_train, axis = -1)
X_test = np.expand_dims(X_test, axis = -1)
```

```python
X_train.shape
```

```python
y_train_label = keras.utils.to_categorical(y_train, class_nums)
y_test_label = keras.utils.to_categorical(y_test, class_nums)
```

```python
y_train[0], y_train_label[0]
```

# Network ANN

```python
model = keras.models.Sequential([
    layers.Flatten(input_shape = [28,28,1]),
    layers.Dense(512, activation = 'relu'),
    layers.Dense(256, activation = 'relu'),
    layers.Dense(class_nums, activation = 'softmax')
])

model.summary();
```

```python
model.compile(optimizer='rmsprop',loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
history = model.fit(X_train, y_train_label, epochs = 10, batch_size = 128, validation_split = 0.2);
```

```python
history_dict = history.history
history_dict.keys()
```

```python
train_loss, val_loss = history_dict['loss'], history_dict['val_loss']
train_acc, val_acc = history_dict['accuracy'], history_dict['val_accuracy']
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize = (15,5))
epoch_runs = [i+1 for i in range(10)]

ax1.plot(epoch_runs, train_loss, 'b', label = 'train_loss')
ax1.plot(epoch_runs, val_loss, 'r', label = 'val_loss', marker = 'o')
ax1.set(title = 'Training and Validation Loss', xlabel = 'Epochs', ylabel = 'Loss')
ax1.legend()
ax2.plot(epoch_runs, train_acc, 'b', label = 'train_accuracy')
ax2.plot(epoch_runs, val_acc, 'r', label = 'val_accuracy', marker = '*')
ax2.set(title = 'Training and Validation Accuracy', xlabel = 'Epochs', ylabel = 'Loss')
ax2.legend()

plt.show();
```

# Model Evaluation

```python
score = model.evaluate(X_test, y_test_label)
```

```python
print(f"Test loss: {score[0]:.4f}")
print(f"Test accuracy: {score[1]:.4f}")
```

```python
from typing import Optional
def plot_data(x_data: np.ndarray, y_data: np.ndarray, y_pred: Optional[np.ndarray]= None):
  nrows,ncols = 2,5
  fig,axes=plt.subplots(nrows,ncols,figsize=(10,4))

  len_x = x_data.shape[0]
  for idx in range(nrows*ncols):
    ax = axes[idx // ncols, idx % ncols]

    img_idx = random.randint(0, len_x)

    ax.imshow(x_data[img_idx], cmap = 'gray')
    ax.set(xticks = [] , yticks = [])
    color = 'green'
    true_label_str = f"True: {class_names[y_data[img_idx]]}"
    if y_pred is not None:
      predicted_idx = np.argmax(y_pred[img_idx])
      predicted_label = class_names[predicted_idx]
      color = 'red' if predicted_idx != y_data[img_idx] else color
      predicted_label_str = f"\nPredicted: {predicted_label}"

    img_title = true_label_str if y_pred is None else true_label_str + predicted_label_str
    ax.set_title(img_title, color = color, fontsize = 10)
  plt.tight_layout()
  plt.show()

plot_data(X_train, y_train)
```
